```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.arima.model import ARIMA
import warnings

warnings.filterwarnings("ignore", category=UserWarning, module="keras")

# Generate Sample Time-Series Data
np.random.seed(42)
time_series_length = 500
num_features = 1  # Adjust if multiple features exist

data = np.sin(np.linspace(0, 50, time_series_length)) + np.random.normal(scale=0.5, size=time_series_length)
df = pd.DataFrame(data, columns=['Value'])
df.index = pd.date_range(start="2002-01-01", periods=time_series_length, freq="W-SAT")  # Explicitly setting frequency

# Normalize Data
scaler = MinMaxScaler(feature_range=(0, 1))
df_scaled = scaler.fit_transform(df)

# Create LSTM Training Dataset
def create_sequences(dataset, look_back):
    X, Y = [], []
    for i in range(len(dataset) - look_back):
        X.append(dataset[i:i + look_back])
        Y.append(dataset[i + look_back])
    return np.array(X), np.array(Y)

look_back = 10  # Number of time steps to look back
X, Y = create_sequences(df_scaled, look_back)

# Split into Training and Testing
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
Y_train, Y_test = Y[:split], Y[split:]

# Define LSTM Model
model = Sequential([
    Input(shape=(look_back, num_features)),  # Correct way to define input shape
    LSTM(50, return_sequences=True),
    LSTM(50),
    Dense(num_features)
])

model.compile(optimizer="adam", loss="mse")

# Train Model
history = model.fit(X_train, Y_train, epochs=50, batch_size=16, validation_data=(X_test, Y_test), verbose=1)

# LSTM Predictions
lstm_train_predicted = model.predict(X_train)
lstm_test_predicted = model.predict(X_test)

# Inverse Transform Predictions
lstm_train_predicted = scaler.inverse_transform(lstm_train_predicted)
lstm_test_predicted = scaler.inverse_transform(lstm_test_predicted)
Y_train_actual = scaler.inverse_transform(Y_train.reshape(-1, num_features))
Y_test_actual = scaler.inverse_transform(Y_test.reshape(-1, num_features))

# ARIMA Predictions
arima_train_predicted = []
arima_test_predicted = []

for i in range(num_features):
    arima_model = ARIMA(Y_train_actual[:, i], order=(5, 1, 0))  # Adjust order as needed
    arima_model_fit = arima_model.fit()
    arima_forecast = arima_model_fit.forecast(steps=len(Y_test_actual))  # Forecasting test set

    arima_train_predicted.append(arima_model_fit.fittedvalues)
    arima_test_predicted.append(arima_forecast)

arima_train_predicted = np.array(arima_train_predicted).T
arima_test_predicted = np.array(arima_test_predicted).T

# Plot Results
plt.figure(figsize=(12, 6))

plt.plot(df.index[split + look_back:], Y_test_actual[:, 0], label="Actual Test Data", color="green")
plt.plot(df.index[split + look_back:], lstm_test_predicted[:, 0], label="LSTM Predictions", color="red")
plt.plot(df.index[split + look_back:], arima_test_predicted[:, 0], label="ARIMA Predictions", color="blue")

plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.title("LSTM vs. ARIMA Time-Series Predictions")
plt.show()
```
